```python
import matplotlib.pyplot as plt
import math
```

```python
def filter_func(x: int, y: int) -> bool:
    #return true if x xor y == x + y
    x_xor_y = x ^ y
    x_plus_y = x + y
    return x_xor_y == x_plus_y
```

```python
start = -1000
end = 1000
step = 1
X = []
Y = []
for x in range(int(start/step), int(end/step)):
    for y in range(int(start/step), int(end/step)):
        if filter_func(x*step, y*step):
            X.append(x*step)
            Y.append(y*step)

plt.scatter(X, Y)
```
